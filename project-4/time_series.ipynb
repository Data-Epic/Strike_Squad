```python
# dependencies
import pandas as pd
import numpy as np
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

warnings.simplefilter('ignore')
%matplotlib inline

pd.pandas.set_option('display.max_columns',None)
```

```python
# variables
CLEANED_DATA_PATH="./data/cleaned_data.parquet"

data = pd.read_parquet(CLEANED_DATA_PATH)
```

```python
df = data.copy()
df.head(3)

df.info()
```

```python
date_column = "DateAcquired"
```

```python
# set the DateAcquired as index to use for time series analysis
df.set_index('DateAcquired', inplace=True)

# Resample data to yearly frequency and plot the number of acquisitions over time
acquisitions_over_time = df.resample('Y').size()
acquisitions_over_time.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Number of Acquisitions Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Acquisitions')
plt.grid(True)
plt.tight_layout()
plt.show()
```

- Based on the analysis, it is observed that the higest rate of the artwork acquisitions were done within the `1965` - `1970` region followed by the `2000` region to `2020` regions
- less impulsive acquisitions were done at a stable rate within the `1930` region to the `1963` region also within `1977` region to the `1999` region

```python
plt.figure(figsize=(12, 6))
df['Width (cm)'].plot(label='Width (cm)', color='green')
df['Height (cm)'].plot(label='Height (cm)', color='blue')
plt.title('Relationship Between Width and Height Over Time (Line Plot)')
plt.xlabel('Date Acquired')
plt.ylabel('Measurement (cm)')
plt.legend()
plt.show()
```

Based on the aquisition dates, we observe works bought on some days had;
- a spike in the size of `widths` of the works within the `1970` region and the heights tend to be at a stable size range
- some noticable and unstble spikes in the size of `widths` of the works within the `1999` region to `2000` region with a stable height size also
- a noticable spike in size of `height` in `2010` and also in `2018` with some slightly unsteady spikes in `widths` too
